Fetch the full timeline for each replier from the seed post.

```python
import polars as pl
from social_llm.scoring.pipeline import collect_network
```

```python
SEED_POST_ID = "REPLACE_ME"

# This fetches all repliers' timelines and saves to data/processed/
posts_df = collect_network(SEED_POST_ID, posts_per_user=100)
```

```python
print(f"{len(posts_df)} total posts from {posts_df['username'].n_unique()} users")
posts_df.head()
```

```python
# Posts per user distribution
posts_per_user = posts_df.group_by("username").len().sort("len", descending=True)
posts_per_user.describe()
```

```python
# Sample some posts
for row in posts_df.sample(min(5, len(posts_df))).to_dicts():
    print(f"@{row['username']}: {row['text'][:300]}")
    print()
```
