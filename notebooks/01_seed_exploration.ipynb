# 01 — Seed Post Exploration

Start from a seed post, pull the conversation, and inspect the repliers.

```python
import polars as pl
from social_llm.threads.client import ThreadsClient
```

```python
# Paste your seed post ID here
SEED_POST_ID = "REPLACE_ME"

client = ThreadsClient()
```

```python
# Fetch the seed post itself
seed = client.get_post(SEED_POST_ID)
print(f"Seed post by @{seed.username}:")
print(seed.text[:500])
```

```python
# Get all replies in the conversation
replies = client.get_conversation(SEED_POST_ID)
print(f"{len(replies)} replies found")
```

```python
# Build a DataFrame of replies
replies_df = pl.DataFrame([r.model_dump() for r in replies])
replies_df.head(10)
```

```python
# Unique repliers
repliers = replies_df.select("username").unique().sort("username")
print(f"{len(repliers)} unique repliers")
repliers
```

```python
# Quick look at reply content — sample a few
for row in replies_df.sample(min(10, len(replies_df))).to_dicts():
    print(f"@{row['username']}: {row['text'][:200]}")
    print()
```
